```python
### n 개의 원소 중 4개 출력하기

# to_pick_list :  뽑아야 되는 숫자들 // 교제에서는 이 부분 따로 선언 X
# picked : 뽑은 숫자들
# pickNum : 뽑아야 되는 숫자들 갯수 ---> Base Case 생성
# 주의할 점 :반복문을 돌릴 때 to_pick_list를 넣으면 에러가 난다 (값이 계속 바뀌기 때문이라고 추정)
# 따라서 새로운 iterater를 지정해주자!


def pick(to_pick_list, picked=[], pickNum=4) :
    

    
    ## Base case : 원소 1개 있을 때
    
    #print('pickNum',pickNum)
    
    if pickNum ==0 :
        print("="*20)
        print("Picked Finished",picked)
        return

    if len(to_pick_list) ==0 :
        return
    
    to_pick_list = sorted(to_pick_list)
    smallest = to_pick_list[0]
    largest = to_pick_list[-1]
    
    
    ## Pick

    for  val in range(smallest,largest+1):
        picked.append(val)
        #print("Appended picked",picked)
        to_pick_list.remove(val)
        #print("Removed to_pick_list",to_pick_list)
        pick(to_pick_list,picked, pickNum -1 )

        picked.remove(val)
        #print("Removed picked",picked,val)

    
```

```python
#test
to_pick_list = [i for i in range(9)]
print(to_pick_list)
pick(to_pick_list)
```
